```python
# TASK 2  Prediction of Store Sales (Python)
# Preprocessing & Feature Engineering
# load and merge Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
train = pd.read_csv("A:/train.csv")
test = pd.read_csv("A:/test.csv")
store = pd.read_csv("A:/store.csv")

train = train.merge(store, on="Store", how="left")
test = test.merge(store, on="Store", how="left")

train["Date"] = pd.to_datetime(train["Date"])
test["Date"] = pd.to_datetime(test["Date"])
store.info()
```

```python
# Date-time Feature Engineering (Extra marks features included)
def create_date_features(df):
    df["Year"] = df["Date"].dt.year
    df["Month"] = df["Date"].dt.month
    df["Week"] = df["Date"].dt.isocalendar().week.astype(int)
    df["Day"] = df["Date"].dt.day
    df["DayOfWeek"] = df["Date"].dt.dayofweek

    df["IsWeekend"] = df["DayOfWeek"].isin([5,6]).astype(int)

    # Beginning / Mid / End of month
    df["MonthPeriod"] = pd.cut(
        df["Day"],
        bins=[0,10,20,31],
        labels=[0,1,2]
    ).astype(int)

    # Holidays
    df["IsHoliday"] = (df["StateHoliday"] != "0").astype(int)

    # Days since last holiday
    df["DaysSinceHoliday"] = (
        df.groupby("Store")["IsHoliday"]
        .apply(lambda x: x.cumsum())
    )

    return df

train = create_date_features(train)
test = create_date_features(test)
```

```python
# Encode Categorical Variables
from sklearn.preprocessing import LabelEncoder

cat_cols = ["StoreType", "Assortment", "PromoInterval"]

for col in cat_cols:
    le = LabelEncoder()
    train[col] = train[col].astype(str)
    test[col] = test[col].astype(str)
    le.fit(pd.concat([train[col], test[col]]))
    train[col] = le.transform(train[col])
    test[col] = le.transform(test[col])
```

```python
# Handle Missing Values
for col in train.columns:
    if train[col].isna().sum() > 0:
        train[col].fillna(train[col].median(), inplace=True)
        test[col].fillna(train[col].median(), inplace=True)
```

```python
# Modeling with sklearn Pipelines
# Feature / Target Split
from sklearn.model_selection import TimeSeriesSplit

# Time-series aware cross validation
tscv = TimeSeriesSplit(n_splits=5)

target = "Sales"
drop_cols = ["Sales", "Customers", "Date", "StateHoliday"]

X = train.drop(columns=drop_cols)
y = train[target]

X_test = test.drop(columns=["Date", "StateHoliday"])

#Train-Validation Split (Time-aware)
from sklearn.model_selection import TimeSeriesSplit

tscv = TimeSeriesSplit(n_splits=5)
```

```python
# Pipeline with Scaling + Random Forest
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", RandomForestRegressor(
        n_estimators=300,
        max_depth=20,
        random_state=42,
        n_jobs=-1
    ))
])

# Loss Function (Defended)

from sklearn.metrics import mean_squared_error

rmse_scores = []

for train_idx, val_idx in tscv.split(X):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    pipeline.fit(X_train, y_train)
    preds = pipeline.predict(X_val)

    rmse = np.sqrt(mean_squared_error(y_val, preds))
    rmse_scores.append(rmse)

print("Mean RMSE:", np.mean(rmse_scores))

# Post-Prediction Analysis
# Feature Importance
pipeline.fit(X, y)

importances = pipeline.named_steps["model"].feature_importances_
feat_imp = pd.Series(importances, index=X.columns).sort_values(ascending=False)

feat_imp.head(15)

# Confidence Interval (Bootstrapping  Extra Marks)
def prediction_interval(model, X, n_boot=100):
    preds = []
    for i in range(n_boot):
        sample = X.sample(frac=1, replace=True)
        preds.append(model.predict(sample))
    return np.percentile(preds, [5, 95], axis=0)

lower, upper = prediction_interval(pipeline, X_test)

# Serialize Models
import joblib
from datetime import datetime

timestamp = datetime.now().strftime("%d-%m-%Y-%H-%M-%S")
model_path = f"sales_model_{timestamp}.pkl"

joblib.dump(pipeline, model_path)

# Deep Learning  LSTM Model
# Time Series Isolation
store_1 = train[train["Store"] == 1].sort_values("Date")
sales_ts = store_1["Sales"].values

# Stationarity Test
from statsmodels.tsa.stattools import adfuller

adf_test = adfuller(sales_ts)
print("ADF p-value:", adf_test[1])

# Differencing
diff_sales = np.diff(sales_ts)

# ACF & PACF
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

plot_acf(diff_sales)
plot_pacf(diff_sales)
plt.show()

# Sliding Window (Supervised)
def create_sequences(data, window=30):
    X, y = [], []
    for i in range(len(data) - window):
        X.append(data[i:i+window])
        y.append(data[i+window])
    return np.array(X), np.array(y)

X_lstm, y_lstm = create_sequences(diff_sales)

# Scaling (-1, 1)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-1,1))
X_lstm = scaler.fit_transform(X_lstm)
y_lstm = scaler.fit_transform(y_lstm.reshape(-1,1))

# LSTM Model (2 Layers)
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

X_lstm = X_lstm.reshape((X_lstm.shape[0], X_lstm.shape[1], 1))

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_lstm.shape[1], 1)),
    LSTM(32),
    Dense(1)
])

model.compile(
    optimizer="adam",
    loss="mse"
)

model.fit(X_lstm, y_lstm, epochs=20, batch_size=32)

# MLflow Model Serving
import mlflow
import mlflow.sklearn

mlflow.start_run()

mlflow.log_metric("RMSE", np.mean(rmse_scores))
mlflow.sklearn.log_model(pipeline, "sales_model")

mlflow.end_run()


# Inference:

loaded_model = mlflow.sklearn.load_model("runs:/<RUN_ID>/sales_model")
predictions = loaded_model.predict(X_test)

```

```python
# Streamlit App
import streamlit as st
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt

# --------------------------------------------------
# Load Model
# --------------------------------------------------
@st.cache_resource
def load_model():
    return joblib.load("sales_model.pkl")

model = load_model()

st.title(" Store Sales Prediction Dashboard")

st.markdown("""
This tool helps **store managers** predict:
-  **Daily Sales**
-  **Number of Customers**
""")

# --------------------------------------------------
# Sidebar Inputs
# --------------------------------------------------
st.sidebar.header("Store Inputs")

store_id = st.sidebar.number_input("Store ID", min_value=1, step=1)
store_type = st.sidebar.selectbox("Store Type", [0,1,2,3])
assortment = st.sidebar.selectbox("Assortment Type", [0,1,2])
competition_distance = st.sidebar.number_input("Competition Distance (meters)", min_value=0)

promo = st.sidebar.selectbox("Promo", [0,1])
promo2 = st.sidebar.selectbox("Promo2", [0,1])

# --------------------------------------------------
# CSV Upload
# --------------------------------------------------
st.subheader(" Upload Input CSV")

uploaded_file = st.file_uploader(
    "Upload CSV with Date-based Features",
    type=["csv"]
)

if uploaded_file:
    df = pd.read_csv(uploaded_file)
    df["Date"] = pd.to_datetime(df["Date"])

    # --------------------------------------------------
    # Feature Engineering (Must match training)
    # --------------------------------------------------
    df["DayOfWeek"] = df["Date"].dt.dayofweek
    df["Month"] = df["Date"].dt.month
    df["Week"] = df["Date"].dt.isocalendar().week.astype(int)
    df["IsWeekend"] = df["DayOfWeek"].isin([5,6]).astype(int)

    # Static features
    df["Store"] = store_id
    df["StoreType"] = store_type
    df["Assortment"] = assortment
    df["CompetitionDistance"] = competition_distance
    df["Promo"] = promo
    df["Promo2"] = promo2

    # Drop unused
    drop_cols = ["Date"]
    X = df.drop(columns=drop_cols)

    # --------------------------------------------------
    # Predictions
    # --------------------------------------------------
    sales_pred = model.predict(X)
    df["Predicted_Sales"] = sales_pred

    # Estimate customers (simple heuristic)
    df["Predicted_Customers"] = (df["Predicted_Sales"] / 10).astype(int)

    # --------------------------------------------------
    # Output
    # --------------------------------------------------
    st.subheader(" Predictions")
    st.dataframe(df[["Predicted_Sales", "Predicted_Customers"]])

    # Plot
    fig, ax = plt.subplots(figsize=(8,4))
    ax.plot(df["Predicted_Sales"], label="Sales")
    ax.plot(df["Predicted_Customers"], label="Customers")
    ax.legend()
    st.pyplot(fig)

    # Download
    csv = df.to_csv(index=False).encode("utf-8")
    st.download_button(
        " Download Predictions",
        csv,
        "sales_predictions.csv",
        "text/csv"
    )
```
